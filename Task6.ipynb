```python
# Forming 5 questions for analysis
questions = [
    "1. How many people have diabetes and how many do not?",
    "2. People of which gender have the most diabetes?",
    "3. What are the common age groups for diabetes cases?",
    "4. People with which BMI range are most prone to diabetes?",
    "5. What are the glucose levels of people with and without diabetes?"
]
questions
```

```python
# Question 1: How many people have diabetes and how many do not?
df['Diagnosis'].value_counts()
```

```python
# Plotting the distribution using bar chart
df['Diagnosis'].value_counts().plot(kind='bar', color=["orchid", "salmon"])
plt.title("Diabetes Cases")
plt.xlabel("1 = Diabetes, 0 = No Diabetes")
plt.ylabel("Count")
plt.show()
```

```python
#plotting a pie chart
df['Diagnosis'].value_counts().plot(kind = "pie", figsize = (8,6))
plt.legend(["Diabetes","No Diabetes"]);
```

```python
#Now let's check how many 'Male' and 'Female' are in the dataset

# 0 represent Female

# 1 represent Male

df.Gender.value_counts()
```

```python
#plotting a pie chart
df.Gender.value_counts().plot(kind = "pie", figsize = (8,6))
plt.title('Male Female ratio')
plt.legend(["Male","Female"]);
```

```python
# Question 2: People of which gender have the most diabetes?
pd.crosstab(df['Diagnosis'], df['Gender'])
```

```python
# Plotting diabetes frequency for each gender
sns.countplot(x='Diagnosis', data=df, hue='Gender')
plt.title("Diabetes Frequency for Gender")
plt.xlabel("0 = No Diabetes, 1 = Diabetes")
plt.show()
```

```python
# Question 3: Common age groups for diabetes cases
df['Age'].value_counts(bins=10)
```

```python
# Plotting distribution of age groups
sns.histplot(df['Age'], bins=10, kde=False)
plt.title("Age Distribution of Diabetes Cases")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Question 4: BMI range prone to diabetes
pd.crosstab(df['Diagnosis'], pd.cut(df['BMI'], bins=[0, 18.5, 24.9, 29.9, 40]))
```

```python
# Using boxplot to show the distribution of BMI for people with and without diabetes
sns.boxplot(x='Diagnosis', y='BMI', data=df)
plt.title("BMI Distribution for Diabetes and Non-Diabetes")
plt.xlabel("0 = No Diabetes, 1 = Diabetes")
plt.ylabel("BMI")
plt.show()
```

```python
# Question 5: Glucose levels for people with and without diabetes
df[['FastingBloodSugar', 'Diagnosis']].groupby('Diagnosis').mean()
```

```python
# Plotting glucose levels for people with and without diabetes
sns.boxplot(x='Diagnosis', y='FastingBloodSugar', data=df)
plt.title("Glucose Levels for Diabetes vs Non-Diabetes")
plt.xlabel("0 = No Diabetes, 1 = Diabetes")
plt.ylabel("Fasting Blood Sugar (mg/dL)")
plt.show()
```
